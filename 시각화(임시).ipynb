#VAYA 워드클라우드

```python
import re
```

```python
#텍스트 전처리
def clean(text) :
  #한국어와 알파벳만 남기고 제거
  #4번 이상 반복되는 문자는 2글자만 남김
  #2칸이상 공백은 1칸만 남김
  pattern1 = re.compile(r"[^ A-Za-z0-9가-힣]+")
  pattern2 = re.compile(r"(.)\1{2,}")
  pattern3 = re.compile(r"\s+")

  text = pattern1.sub("", str(text))
  text = pattern2.sub(r"\1" * 2, str(text))
  text = pattern3.sub(" ", str(text))

  return text.strip()
```

```python
celeb_VAYA['comments'] = [clean(c) for c in celeb_VAYA['comments']]
celeb_VAYA
```

```python
#label 별로 분류
nega = celeb_VAYA[celeb_VAYA['label']==0]['comments']
norm = celeb_VAYA[celeb_VAYA['label']==1]['comments']
posi = celeb_VAYA[celeb_VAYA['label']==2]['comments']
```

```python
from PyKomoran import *
```

```python
komoran = Komoran("STABLE")
```

```python
def tokenize(text) :
  result = []
  stop_pos_tags = ['IC', 'JKS', 'JKC', 'JKG', 'JKO',
                 'JKB', 'JKV', 'JKQ', 'JX', 'EF', 'ETN',
                 'ETM', 'XSA', 'SF', 'SP', 'SS', 'SE', 'SO',
                 'SL', 'SH', 'SW', 'NF', 'NV', 'SN', 'NA']

  pairs = komoran.get_list(text)
  for pair in pairs :
    morph = pair.get_first()
    pos = pair.get_second
    if pos not in stop_pos_tags:
      if pos in ['VV', 'VA', 'VX', 'VCP', 'VCN']:
        morph = morph+'다'

      result.append(morph)
  
  noun = (komoran.get_nouns(text))
  if pos not in stop_pos_tags:
    if pos in ['VV', 'VA', 'VX', 'VCP', 'VCN'] :
      noun = noun+'(시)다'

  return noun
```

```python
komo_posi = tokenize(posi_txt)
```

```python
komo_posi
```

```python
#2글자 이상 추출
komo_new = []
for word in komo_posi :
  if len(word) >= 2:
    komo_new.append(word)
```

```python
#단어 count
from collections import Counter
cnt = Counter(komo_new)
```

```python
sorted_dict = sorted(cnt.items(), key = lambda item: item[1], reverse = True)
print(sorted_dict)
```
